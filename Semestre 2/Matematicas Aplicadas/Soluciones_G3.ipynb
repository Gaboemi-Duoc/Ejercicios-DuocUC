# Problema 1

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

## Solución

**1)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

def r(x):
  return -1.5 * x**2 + 11.5 * x - 15

def g(x):
  return 1.58 * x**4 - 19.17 * x**3 + 80.92 * x**2 - 139.33 * x + 85

x = np.arange(0, 50, 0.01)

plt.plot(x, g(x), color = 'r', label = 'g(x)')
plt.plot(x, r(x), color = 'g', label = 'r(x)')
plt.ylim(-10, 10)
plt.xlim(0, 6)

plt.title('Comparación entre las funciones g(x) y r(x)')

plt.grid(True)
plt.legend()
plt.show()
```

**2)**

La intersección ocurre cuando:

\begin{split}
r(x) & = g(x)\\
0 & = g(x)-  r(x)
\end{split}

```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def inter(x):
  return g(x)-r(x)

x = np.linspace(0, 15, 15)

#np.around() permite redondear los valores a 2 decimales y elimina los restantes
solucion = np.around(fsolve(inter, x), decimals = 2)

#np.unique() permite eliminar las soluciones repetidas
interseccion = np.unique(solucion)

print(f'Los puntos de intersección de las funciones están en x = {interseccion}')
```

# Problema 2

Sean las funciones:

\begin{split}
h(x) & = 2(x - 2)^{2}\\
i(x) & = x + 1
\end{split}

<br>

1. Grafique ambas funciones en un mismo gráfico utilizando *Matplotlib*.
2. ¿En qué punto(s) interseca la función $h(x)$ al eje $x$? ¿y la función $i(x)$?
3. ¿Las gráficas de las funciones se intersecan en algún punto?
4. ¿Para qué valores de $x$ ocurre que $h(x) \leq i(x)$?
5. Determine el punto en que $h(x) = 50$.


## Solución

**1)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

def h(x):
  return 2 * (x - 2)**2

def i(x):
  return x + 1

x = np.arange(-5, 100, 0.01)

plt.plot(x, h(x), color = 'm', label = 'h(x)')
plt.plot(x, i(x), color = 'y', label = 'i(x)')
plt.ylim(-1, 10)
plt.xlim(-2, 5)

plt.title('Comparación entre las funciones h(x) y i(x)')

plt.grid(True)
plt.legend()
plt.show()
```

**2)**

Intersección de $h(x)$ con el eje $x$:

$0 = h(x)$

Intersección de $i(x)$ con el eje $x$:

$0 = i(x)$

```python
import numpy as np
from scipy.optimize import fsolve

x = np.linspace(0, 15, 1)

#np.around() permite redondear los valores a 2 decimales y elimina los restantes
solucion_h = fsolve(h, x)
solucion_i = fsolve(i, x)

print(f'La función h(x) interseca al eje x en x = {solucion_h[0]:.0f} y la función i(x) en x ={solucion_i[0]:.0f}.')
```

La función $h(x)$ interseca al eje x en el punto $(2, 0)$ y la función $i(x)$ en el punto $(-1, 0)$.

**3)**
La intersección ocurre cuando:

\begin{split}
h(x) & = i(x)\\
0 & = i(x)-  h(x)
\end{split}

```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def inter(x):
  return i(x)-h(x)

x = np.linspace(0, 15, 15)

#np.around() permite redondear los valores a 2 decimales y elimina los restantes
solucion = np.around(fsolve(inter, x), decimals = 2)

#np.unique() permite eliminar las soluciones repetidas
interseccion = np.unique(solucion)

print(f'Los puntos de intersección de las funciones están en x = {interseccion}.')
```

```python
#Calculando los valores de las imagenes
print(f'h(1) = {h(1):.2f}')
print(f'h(3,5) = {h(3.5):.2f}')
```

La función $h(x)$ interseca a la función $i(x)$ en el punto $(1, 2)$ y en el punto $(3,5; 4,5)$.

**4)**

Observando el gráfico y analizando las soluciones de ejercicio anterior se puede concluir que, para valores de $x$ mayores a $1$ pero menores a $3,5$ la función $h(x)$ es menor a $i(x)$.

**5)**

$50 = h(x)$

$0 = h(x)- 50$

```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def cincuenta(x):
  return h(x) - 50

x = np.linspace(-200, 200, 50)

#np.around() permite redondear los valores a 2 decimales y elimina los restantes
solucion = np.around(fsolve(cincuenta, x), decimals = 0)

#np.unique() permite eliminar las soluciones repetidas
interseccion = np.unique(solucion)

print(f'La función h(x) es igual a 50 en x = {interseccion}.')
```

## Solución

**1)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

def w(x):
  return 0.6 * x**2 + 1.5

def t(x):
  return 0.8 * x**3 - 10 * x + 10

x = np.arange(-10, 10, 0.1)

plt.plot(x, w(x), color = 'm', label = 'w(x)')
plt.plot(x, t(x), color = 'y', label = 't(x)')
plt.ylim(-4, 24)
plt.xlim(-3, 4)

plt.title('Comparación entre las funciones w(x) y t(x)')

plt.grid(True)
plt.legend()
plt.show()
```

**2)**

La intersección ocurre cuando:

\begin{split}
w(x) & = t(x)\\
0  & = t(x)-  w(x)
\end{split}

```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(x):
  return w(x)- t(x)

x = np.linspace(-4, 4, 4)

#np.around() permite redondear los valores a 2 decimales y elimina los restantes
solucion = np.around(fsolve(interseccion, x), decimals = 2)

#np.unique() permite eliminar las soluciones repetidas
unica = np.unique(solucion)

print(f'Los puntos de intersección de las funciones están en x = {unica}.')
```

Intervalo $(0,86;3,47)$.

**3)**

# Problema 5

Te contrataron como asesor en una campaña de marketing digital y necesitas equilibrar el ingreso por ventas con los costos de inversión.

Las funciones de ingresos por ventas $I(x)$ y costos de inversión $C(x)$ (ambas en miles de pesos) son:

<br>

$$I(x) = - 20x + 1.000$$

$$C(x) = 200x + 500$$

<br>

Donde $x$ corresponde al tiempo transcurrido desde que se inició la campaña (en semanas).

<br>

1. Analice el comportamiento del ingreso por ventas y del costo de inversión a medida que aumenta el tiempo desde que se inició la campaña. Considere la pendiente y el intercepto de cada función para realizar el análisis.
2. Grafique ambas funciones en un mismo gráfico, indicando nombre
de los ejes. Utilice *Matplotlib*.
3. Determine, utilizando *Python*, el punto de equilibrio entre ambas funciones e interprete el resultado.
4. Utilizando el gráfico y lo realizado en el punto $3$, ¿qué puedes concluir? ¿Aconsejarías mantener la campaña transcurridas $5$ semanas desde el inicio de la campaña?

## Solución

**1)**

La función $I(x)$ indica que al inicio de la campaña el ingreso por ventas es de $\$1.000.000$. A medida que las semanas aumentan, el ingreso por ventas disminuye. Específicamente, por cada semana de campaña, el ingreso por ventas disminuye en $20$ mil pesos.

La función $C(x)$ indica que al inicio de la campaña  el costo por inversión es de $\$500.000$. A medida que las semanas aumentan, el costo por inversión aumenta. Específicamente, por cada semana de campaña, el costo de inversión aumenta en $200$ mil pesos.

**2)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

#Ingresos por ventas
def I(x):
  return -20 * x + 1000

#Costos de inversión
def C(x):
  return 200 * x + 500

x = np.arange(0.1, 200, 0.1)

plt.plot(x, I(x), color = 'g', label = 'Ingresos por venta')
plt.plot(x, C(x), color = 'r', label = 'Costos de inversión')
plt.ylim(500, 1250)
plt.xlim(0, 7)

plt.title('Relación entre los ingresos y costos de inversión en función del tiempo')
plt.xlabel('Tiempo transcurrido (semanas)')
plt.ylabel('Ingresos y Costos de inversión (miles de pesos)')

plt.grid(True)
plt.legend()
plt.show()
```

**3)**

El punto de equilibrio ocurre cuando:

\begin{split}
I(x) & = C(x)\\
0 & = C(x) - I(x)
\end{split}


```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(x):
  return C(x) - I(x)

x = np.linspace(0, 20, 5)
solucion = np.around(fsolve(interseccion, x), decimals = 1)

semanas = np.unique(solucion)

print(f'El punto de equilibrio ocurre en x = {semanas}.')
print(f'Cuando han transcurrido {semanas[0]} semanas, los ingresos por ventas igualan a los costos de inversión: {I(semanas[0]):.0f} miles de pesos.')
```

**4)**

Aconsejaría no continuar con la campaña, ya que transcurridas más de $2,3$ semanas los costos de inversión superan a los ingresos por ventas.

# Problema 7

Te contrataron para realizar una optimización de una plataforma de streaming. Debes determinar donde se equilibran los ingresos por suscripciones con los costos de operación

Las funciones de ingresos por suscriptores $I(x)$ y costos de operación $C(x)$ (ambos en miles de pesos) son:

<br>

$$I(x) = 10x^{2} + 50x$$

$$C(x) = 5x^{2} + 80x + 100$$

<br>

Donde $x$ corresponde al número de suscriptores (en miles).

<br>

1. Debes determinar dónde se equilibran los ingresos por suscripciones con los costos de operación.
2. Analice en qué tramo los costos de operación son mayores que los ingresos por suscriptores.

## Solución

**1)**

El punto de equilibrio ocurre cuando:

\begin{split}
I(x) & = C(x)\\
0 & = I(x) - C(x)
\end{split}


```python
import numpy as np
from scipy.optimize import fsolve

#Ingreso por suscriptores
def I(x):
  return 10 * x**2 + 50 * x

#Costos de operación
def C(x):
  return 5 * x**2 + 80 * x + 100

#Definiendo la función igualada a cero
def interseccion(x):
  return I(x) - C(x)

x = np.linspace(0, 10, 2)
solucion = np.around(fsolve(interseccion, x), decimals = 1)
suscriptores = np.unique(solucion)

suscriptor = [elementos for elementos in suscriptores if elementos >= 0]

print(f'El punto de equilibrio ocurre en x = {suscriptores}.')
print(f'Cuando la plataforma tiene {suscriptor[0]} mil suscriptores, el ingreso por suscriptores es igual a los costos de operación: {I(suscriptor[0]):.2f} mil pesos.')
```

**2)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

#Ingreso por suscriptores
def I(x):
  return 10 * x**2 + 50 * x

#Costos de operación
def C(x):
  return 5 * x**2 + 80 * x + 100

x = np.arange(0.1, 200, 0.1)

plt.plot(x, I(x), color = 'r', label = 'Ingresos por suscripción')
plt.plot(x, C(x), color = 'g', label = 'Costos de operación')
plt.ylim(0, 4000)
plt.xlim(0, 20)

plt.title('Relación entre los Ingresos y Costos en función de la cantidad de suscriptores')
plt.xlabel('Cantidad de suscriptores (miles de unidades)')
plt.ylabel('Ingresos y Costos (miles de pesos)')

plt.grid(True)
plt.legend()
plt.show()
```

Cuando la plataforma de streaming tiene menos de $8.400$ suscriptores, los costos de operación superan a los ingresos por suscriptores.

# Problema 8

Constanza es agente inmobiliario, el sueldo en su actual trabajo depende de la cantidad de casas vendidas, le pagan $\$600.000$ de sueldo base, más una comisión del $5\%$ por cada casa vendida. Lo anterior puede modelarse mediante la función $S_{actual}(x)$ (en miles de pesos):

<br>

$$S_{actual}(x) = 600 \cdot 1,05^{x}$$

<br>

donde $x$ corresponde a la cantidad de casas vendidas durante el mes. Debido al aumento en el costo de la vida, necesita aumentar sus ingresos. Luego de postular a varios trabajos, recibió una oferta laboral donde le ofrecen un sueldo base de $\$500.000$ más una comisión de venta del $7\%$ por cada casa vendida.

<br>

Considerando el contexto anterior encuentre:

1. La forma algebraica que modela el sueldo mensual de la oferta recibida.
2. Grafique ambas funciones usando *Matplotlib*. Considera que vende un máximo de $30$ casas mensuales.
3. Si tuvieras que aconsejar a Constanza, ¿le recomendarías que tome la oferta o que se mantenga en su actual trabajo? Justifica matemáticamente.

## Solución

**1)**

La forma algebraica que modela el sueldo de la oferta es:

$$S_{oferta}(x) = 500\cdot 1,07^{x}$$

**2)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

#Sueldo actual
def S_actual(x):
  return 600 * 1.05**x

#Sueldo oferta
def S_oferta(x):
  return 500 * 1.07**x

x = np.arange(0, 50, 1)

plt.plot(x, S_actual(x), color = 'r', label = 'Sueldo actual')
plt.plot(x, S_oferta(x), color = 'g', label = 'Sueldo oferta')
plt.ylim(400, 2500)
plt.xlim(0, 35)

plt.title('Relación entre los sueldos y la cantidad de casas vendidas')
plt.xlabel('Cantidad de casas vendidas (unidades)')
plt.ylabel('Sueldo actual y sueldo de la oferta (miles de pesos)')

plt.grid(True)
plt.legend()
plt.show()
```

**3)**

El punto de equilibrio ocurre cuando:

\begin{split}
S_{actual}(x) & = S_{oferta}(x)\\
0 & = S_{oferta}(x) - S_{actual}(x)
\end{split}

$$f(x) = \sqrt{500 x + 2.000}$$

$$g(x) = \sqrt{800 x + 1.000}$$

donde $x$ corresponde a la intensidad de la luz (en lux).

## Solución

**1)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

#Primera especie de algas
def f(x):
  return np.sqrt(500 * x + 2000)

#Segunda especie de algas
def g(x):
  return np.sqrt(800 * x + 1000)

x = np.arange(0, 500, 1)

plt.plot(x, f(x), color = 'm', label = 'Primera especie: f(x)')
plt.plot(x, g(x), color = 'y', label = 'Segunda especie: g(x)')
plt.ylim(0, 150)
plt.xlim(0, 20)

plt.title('Relación entre la tasa de crecimiento de dos especies de algas y la intensidad de luz recibida')
plt.xlabel('Intensidad de luz (lux)')
plt.ylabel('Tasa de crecimiento (mm/día)')

plt.grid(True)
plt.legend()
plt.show()
```

El punto de equilibrio ocurre cuando:

\begin{split}
f(x) & = g(x)\\
0 & = g(x) - f(x)
\end{split}


```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(x):
  return g(x) - f(x)

x = np.linspace(0, 10, 2)
solucion = np.around(fsolve(interseccion, x), decimals = 2)
intensidad = np.unique(solucion)

print(f'El punto de intersección ocurre en x = {intensidad}.')
print(f'Cuando la intensidad de la luz es {intensidad[0]} lux, la tasa de crecimiento es igual en ambas especies: {f(intensidad[0]):.2f} mm/día.')
```

# Problema 10

En una planta de tratamiento de aguas residuales, los ingenieros están estudiando la eficiencia, medida en porcentaje, de dos diferentes procesos de filtración para eliminar contaminantes. La eficiencia de cada proceso varía en función del tiempo de tratamiento, $t$, medido en horas.

Las funciones que describen la eficiencia de los procesos son las siguientes:

Proceso A:

 $$A(t) = 1,4t^{4} - 2,55t^{3} - 3,37t^{2} + 2,75t + 10$$

Proceso B:

 $$B(t) = t^{4} -2,67t^{3} - t^{2} + 2t + 20$$

 <br>

1. Grafique ambas funciones indicando el nombre de cada eje coordenado, junto con su unidad de medida. Utilice *Matplotlib*.
2. Transcurridas $3$ horas y media, ¿qué porcentaje de eficiencia tiene el proceso A? ¿Y el B?
3. ¿Para cuántas horas transcurridas el proceso A tiene la misma eficiencia que el proceso B? Indique la eficiencia en este punto.
4. Si sabe que los procesos demoran como máximo $5$ horas, analice cuándo es conveniente aplicar el proceso A.

## Solución

**1)** Gráfico de las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

#Proceso A
def A(t):
  return 1.4 * t**4 - 2.55 * t**3 - 3.37 * t**2 + 2.75 * t + 10

#Proceso B
def B(t):
  return t**4 - 2.67 * t**3 - t**2 + 2 * t + 20

t = np.arange(-100, 5000, 0.1)

plt.plot(t, A(t), color = 'lime', label = 'Proceso A')
plt.plot(t, B(t), color = 'deeppink', label = 'Proceso B')
plt.ylim(0, 150)
plt.xlim(-5,5)

plt.title('Relación entre la eficiencia de dos procesos y el tiempo transcurrido')
plt.xlabel('Tiempo transcurrido (horas)')
plt.ylabel('Eficiencia del proceso (porcentaje)')

plt.grid(True)
plt.legend()
plt.show()
```

**2)**


```python
print(f'Transcurridas 3,5 horas, el proceso A tiene una eficiencia del {A(3.5):.1f}% y el B de {B(3.5):.1f}%.')
```

**3)**

La intersección ocurre cuando:

\begin{split}
A(t) & = B(t)\\
0 & = B(t) - A(t)
\end{split}




```python
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(t):
  return B(t) - A(t)

t = np.linspace(-10, 10, 4)
solucion = np.around(fsolve(interseccion, t), decimals = 1)
unica = np.unique(solucion)

tiempo = [elementos for elementos in unica if elementos >= 0]

print(f'El punto de intersección ocurre en t = {tiempo}.')
print(f'Cuando han transcurrido {tiempo[0]} horas, ambos procesos muestran la misma eficiencia: {A(tiempo[0]):.1f}%.  ')
```

**4)**

El proceso A es más eficiente si se sabe que el tratamiento debe durar más de $2,8$ horas. Para tratamientos que duren menos de $2,8$ horas es conveniente usar el proceso B.

## Solución

**1)** Gráfico de la función

```python
import matplotlib.pyplot as plt
import numpy as np

def tramo_1(x):
  return 2 - x

def tramo_2(x):
  return 3

#Vectorizamos el tramo 2 para que tenga las mismas dimensiones que "x"
tramo_2 = np.vectorize(tramo_2)

x = np.arange(-50, 50, 0.001)
f_x = np.piecewise(x, [x < 1, x >= 1], [lambda x: tramo_1(x), lambda x: tramo_2(x)])


plt.plot(x[x < 1], tramo_1(x[x < 1]), color = 'darkred', label = 'Tramo 1')
plt.plot(x[x >= 1], tramo_2(x[x >= 1]), color = 'c', label = 'Tramo 2')

plt.title('Gráfico de f(x)')

plt.grid(True)
plt.legend()
plt.show()
```

Para valores de $x$ menores a $1$, la función es lineal decreciente y para valores de $x$ iguales o mayores a $1$, la función es constante.

**2)**

```python
def f(x):
  if x < 1:
    f = 2 - x
  else:
    f = 3
  return f

#Evaluando en la función
print(f'Cuando x = -5, f(x) = {f(-5)}. \nCuando x = 5, f(x) = {f(5)}.')
```
